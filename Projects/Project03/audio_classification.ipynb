# Audio classification

```python
import requests
import json
```

```python
fileName = "data/first_sound.wav"

url = "http://myibmcontainer.eastus.azurecontainer.io:5000/model/predict" # this is a server I set up for the class
files = {"audio":open(fileName, 'rb')}
r = requests.post(url, files=files) # This is the same requests as the web scraping, but now we post rather than get
```

```python
json.loads(r.text)
```
